```r
# Crear una lista para almacenar los datos de cada año
datos_por_año <- list()

# Iterar sobre los años
for (año in 1971:2015) {
  # Obtener los datos para el año actual
  datos <- sum_CE_pr0.1_tmax30[[as.character(año)]]
  
  # Crear una variable para el año actual
  variable_nombre <- paste0("y", año, "_num")
  
  # Asignar los datos al nombre de la variable correspondiente
  assign(variable_nombre, bindGrid(datos$Winter, datos$Spring, datos$Summer, datos$Autumn, dimension= "member", skip.temporal.check = TRUE))
  
  # Almacenar los datos en la lista
  datos_por_año[[as.character(año)]] <- get(variable_nombre)
  print(variable_nombre)
  spatialPlot(variable_nombre, backdrop.theme = "coastline", color.theme = "jet.colors", names.attr = c("Winter", "Spring", "Summer", "Autumn"), main = paste0(año, " Combined Events"), set.min = 0, set.max = 100, at = seq(0,100), as.table = TRUE)
}

# Mostrar el gráfico para un año específico (por ejemplo, 1982)
# spatialPlot(y1982_num, backdrop.theme = "coastline", color.theme = "jet.colors", names.attr = c("Winter", "Spring", "Summer", "Autumn"), main = "1982 Combined Events", set.min = 0, set.max = 100, at = seq(0,100), as.table = TRUE)
```
